Backesting Algorithm 

```python
def predict(train ,test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds= model.predict(test[predictors])
    preds= pd.Series(preds,index=test.index,name="Predictions")
    combined = pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions= []
    for i in range(start, data.shape[0],step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors , model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []
for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column]=sp500["Close"]/rolling_averages["Close"]
    trend_column = f"Trend_{horizon}"
    sp500[trend_column]= sp500.shift(1).rolling(horizon).sum()["Target"]
    new_predictors+= [ratio_column,trend_column]
```

```python
sp500= sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50 , random_state=1)
```

```python
def predict(train ,test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds= model.predict_proba(test[predictors])[:,1]
    preds[preds>=.6]=1
    preds[preds<.6]=0
    preds= pd.Series(preds,index=test.index,name="Predictions")
    combined = pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
predictions = backtest(sp500,model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
for horizon in horizons:
    # Calculate rolling mean for Volume
    volume_rolling = sp500["Volume"].rolling(horizon).mean()
    
    # Create the new Volume ratio column
    volume_ratio_column = f"Volume_Ratio_{horizon}"
    
    # Use .loc to safely assign the new column
    sp500.loc[:, volume_ratio_column] = sp500["Volume"] / volume_rolling

    # Add the new predictor to the list
    new_predictors.append(volume_ratio_column)
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
from sklearn.metrics import precision_score, confusion_matrix, classification_report

# Run the backtest with the updated predictors including Volume-based features
predictions = backtest(sp500, model, new_predictors)

# Display the counts of predictions (0s and 1s)
print("Prediction Counts:\n", predictions["Predictions"].value_counts())

# Calculate and print the precision score
precision = precision_score(predictions["Target"], predictions["Predictions"])
print(f"Precision Score: {precision:.4f}")

# Display the confusion matrix to assess model performance
conf_matrix = confusion_matrix(predictions["Target"], predictions["Predictions"])
print("Confusion Matrix:\n", conf_matrix)

# Generate a detailed classification report for further insights
class_report = classification_report(predictions["Target"], predictions["Predictions"])
print("Classification Report:\n", class_report)

# Visualize the predictions vs actual values
predictions.plot(title="Predictions vs Actual Values", figsize=(10, 6))
```
